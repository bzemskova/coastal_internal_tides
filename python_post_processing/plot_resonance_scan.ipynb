### This notebook shows how to open the Matlab file containing resonance scan data, pressure response of the solution sweeping over forcing frequencies $\omega$ and along-shore wavenumber $k$,  and plotting this resonance response.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

Open Matfile using loadmat module and extract all the data from the loaded dictionary.

```python
data = loadmat('../data/resonance_scan.mat')
set_up_data = loadmat('../data/example_data_600_200.mat')
```

Some of these variables will not be used for plotting resonance scan, but they will be useful to keep in mind what the simulation set-up was.

```python
#SET-UP VARIABLES
L = data['L'] #domain size (m): size[1,]
Nx = data['Nx'] #number of grid points in x: size[1,]
Nz = data['Nz'] #number of grid points in z: size[1,]
R = data['R'] #surface boundary condition (=0: rigid lid, =1: free surface): size[1,]
f = data['f'] #Coriolis parameter (1/s): size[1,]
W = data['W'] #slope width (m): size[1,]
xs = data['xs'] #shelf width (m): size[1,]
g = data['g'] #gravity (m/s^2): size[1,]
h0 = data['h0'] #ocean depth (m): size[1,]
hc= data['hc'] #depth at the coast (m): size[1,]
hs= data['hs'] #depth at the shelf break (m): size[1,]
Zpyc= data['Zpyc'] #pycnocline depth (m): size[1,] --> negative value
mupyc= data['mupyc'] #pycnocline width (m): size[1,]
xx = set_up_data['xx'] #grid in x
z = set_up_data['z'] #grid in z
omega_vec = data['SI'] #array of frequencies (1/s): size[sx,]
k_vec = data['l_vec2'] #array of along-shore wavenumbers (1/m): size[kx,]
```

```python
#PRODUCED VARIABLES
h = data['h'] #topographic depth (m): size[Nx,Nz]
hx = data['hx'] #topographic slope (dh/dx): size[Nx,Nz]
N2 = set_up_data['N2'] #stratification (1/s^2): size[Nx,Nz]
P0p = data['P0p_sd'] #pressure response: size[sx,kx]
```

**We will plot the set-up of this problem first. Color will indicate stratification $N^2\, (s^{-2})$ The specifications are as follows:** 
- **topography parameters:** shelf width $x_s=80$km, slope width $W=32$km, domain width $L=800$km, deep ocean depth $h_0=3100$m, depth at the coast $h_c=100$m, depth at the shelf break $h_s=150$m;
- **stratification parameters:** pycnocline depth $Z_{pyc}=600$m, pycnocline width $\mu_{pyc}=200$m.

```python
fig, ax0 = plt.subplots()
im = ax0.pcolormesh(xx/1e3,z,N2,shading='gouraud')
fig.colorbar(im, ax=ax0)
ax0.set_facecolor('gray')
ax0.set_title('Stratification $N^2 (s^{-2})$')
ax0.set_xlabel('Cross-shore distance x (km)')
ax0.set_ylabel('Depth z (m)');
```

We will plot the entire resonance scan here: darker lines indicate ($\omega$,k) pairs where the pressure response is amplified, so there is resonance, i.e., modes. \
The red line indicates inertial frequency $\omega=f$. We can see modes transitioning from subinertial to superinertial regime across this line.

```python
fig, ax0 = plt.subplots()
omega_f_vec = omega_vec/f
l_vec_norm = k_vec/(1e-4)
k = l_vec_norm.reshape(k_vec_norm.shape[1],)
omega = np.linspace(0,omega_vec.max(),k_vec_norm.shape[1])

#Plot pressure response as a function of along-shore wavenumber (k) and frequency (\omega/f)
im = ax0.pcolormesh(k, (omega_f_vec.reshape(omega_f_vec.shape[1],)), 
            np.log10(P0p), cmap=plt.get_cmap('gray_r'),shading='gouraud',
            vmin=-10, vmax=3)
fig.colorbar(im, ax=ax0)

#Add line for inertial frequency \omega=f
ax0.plot(k,np.ones(k_vec_norm.shape[1],),'r',linewidth=2)

ax0.set_title('Pressure response')
ax0.set_xlabel('Along-shore wavenumber k ($10^{-4}\,m^{-1}$)')
ax0.set_ylabel('$\omega/f$')
ax0.set_xlim((-1.5, 1.5));
```

We can also zoom in on the region close to $l=0$ (barotropic mode). We can plot (in blue) the dispersion curves for free-surface Kelvin waves on the shelf: $l=\omega/\sqrt{gh_s}$; for free surface Kelvin waves in the deep ocean: $l=\omega/\sqrt{gh_0}$; and Poincare waves: $l=\sqrt{(\omega^2-f^2)/gh_s}$.

```python
fig, ax0 = plt.subplots()
#Plot pressure response as a function of along-shore wavenumber (k) and frequency (\omega/f)
im = ax0.pcolormesh(k, (omega_f_vec.reshape(omega_f_vec.shape[1],)), 
            np.log10(P0p), cmap=plt.get_cmap('gray_r'),shading='gouraud',
            vmin=-10, vmax=3)
fig.colorbar(im, ax=ax0)

#Add line for inertial frequency \omega=f
ax0.plot(k,np.ones(k_vec_norm.shape[1],),'r',linewidth=2)

#Add line for Kelvin wave dispersion on the shelf: k = /omega/\sqrt(g hs)
ax0.plot( omega/np.sqrt(g.reshape(1,)*hc.reshape(1,))/(1e-4),omega/f.reshape(1,), ':b', linewidth=1.5)
ax0.plot( -omega/np.sqrt(g.reshape(1,)*hc.reshape(1,))/(1e-4),omega/f.reshape(1,), ':b', linewidth=1.5)

#Add line for Kelvin wave dispersion in the deep ocean: k = /omega/\sqrt(g h0)
ax0.plot( omega/np.sqrt(g.reshape(1,)*h0.reshape(1,))/(1e-4),omega/f.reshape(1,), ':g', linewidth=1.5)
ax0.plot( -omega/np.sqrt(g.reshape(1,)*h0.reshape(1,))/(1e-4),omega/f.reshape(1,), ':g', linewidth=1.5)

#Plot Pointcare wave dispersion
oo = np.linspace(f.reshape(1,),omega_vec.max(),k_vec_norm.shape[1]).reshape(k_vec_norm.shape[1],)
ll = np.sqrt((oo**2 - f.reshape(1,)**2)/(g.reshape(1,)*hc.reshape(1,)))
mm = np.argwhere(np.real(ll)>0)
ax0.plot(ll[mm]/1e-4,oo[mm]/f.reshape(1,),'m:', linewidth=1.5) ;
ax0.plot(-ll[mm]/1e-4,oo[mm]/f.reshape(1,),'m:', linewidth=1.5) ;
#for ii in range(1,4):
#    scl = 4/(2*ii - 1) ;
#    kk = 2*np.pi/(scl*1000e3) ;
#    - kk**2 ) ; #
#    mm = np.argwhere(np.real(ll)>0)
#    ax0.plot(ll[mm]/1e-4,oo[mm]/f.reshape(1,),'b--') ;
#    ax0.plot(-ll[mm]/1e-4,oo[mm]/f.reshape(1,),'b--') ;



ax0.set_title('Pressure response')
ax0.set_xlabel('Along-shore wavenumber k ($m^{-1}$)')
ax0.set_ylabel('$\omega/f$')
ax0.set_xlim((-0.2, 0.2))
ax0.set_ylim((0.75,2))
```
